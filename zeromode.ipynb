```python
import numpy as np
from matplotlib import pyplot as plt
from numpy import *
import math
import pylab
import matplotlib.pyplot as plt
from scipy.integrate import trapz
from scipy.integrate import quad
from scipy.integrate import simps
plt.rcParams["figure.figsize"] = [7.50, 3.50]
plt.rcParams["figure.autolayout"] = True

  
L=1;  #value for L
e=1; #value for $\epsilon$
m=1;   #value for m
def f(k2):
   return L/(4*2*3.14*math.sqrt(2*e*L))*1/(k2*k2*L/(2*e)+1)**(3/2)
def g(k):
   return 1/(4*math.sqrt(2*e*L))*1/(k*k*L/(2*e)+1)**(3/2)
def s(n):
  return (1)/(4*(pow(2*e*L,1/2)))*(pow((((pow(abs(n)*2*pi,2)))/(2*e*L)+(pow(m,2))),-3/2))
sigma = 1/(4*np.pi*m**2)

print("L =", L)
print("ε =", e)
print("m =", m)
print("-----------------")
print("Σ = i/4π(m^2) = ", sigma,"i")
print("-----------------")
k = np.linspace(-5, 5,11)
k2 = np.linspace(-5, 5,11)
n = np.linspace(-5, 5,11)
Sum = np.sum(s(n))
Integration, err = quad(f,-500,500)
print("Summation =", Sum,"i")
print("Integral =", Integration,"i")
errorsum = abs(Sum-Integration)*100/(Integration)
print("Error in Sum from Integral =", errorsum, "%")
print("-----------------")
plt.plot(k2, f(k2), color='red', label='Integrand')
#plt.plot(k, g(k), color='blue')
plt.bar(k, g(k),width = 1, label='Summation Expression')
plt.xlabel("$k^+$")
plt.ylabel("$\Sigma(k^+)$")
plt.title("Summation Expression vs. Integrand")
plt.legend(loc="upper right")
#plt.savefig('filename.png', dpi=500)
#plt.ylim([0, 0.0000051])
plt.show()
```
